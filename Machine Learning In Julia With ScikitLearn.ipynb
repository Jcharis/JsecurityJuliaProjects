```julia
Ylabels = iris[:species]
```

```julia
Xfeatures = convert(Array,iris[[:sepal_length,:sepal_width,:petal_length,:petal_width]])
```

```julia
Ylabels = convert(Array, iris[:species])
```

### Is the problem Classification or Regression?

```julia
using ScikitLearn: linear_model
```

```julia
@sk_import linear_model: LogisticRegression
#@sk_import tree: DecisionTreeClassifier
```

```julia
model = LogisticRegression(fit_intercept=true)
fit!(model, Xfeatures, Ylabels)

```

```julia
accuracy = sum(predict(model, Xfeatures) .== Ylabels) / length(Ylabels)
println("Accuracy of model: $accuracy")
```

```julia
using ScikitLearn.CrossValidation: cross_val_score

cross_val_score(LogisticRegression(), Xfeatures, Ylabels; cv=3) 
```
